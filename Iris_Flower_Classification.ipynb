```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
iris = pd.read_csv("iris.csv")
iris.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Split into features and labels
X = iris.drop('Species', axis=1)
y = iris['Species']
```

```python
# Split data into train and test
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Feature Scaling
scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
test_X_scaled = scaler.transform(test_X)

# Polynomial Features
poly = PolynomialFeatures(degree=2, include_bias=False)
train_X_poly = poly.fit_transform(train_X_scaled)
test_X_poly = poly.transform(test_X_scaled)
```

```python
# 1. Support Vector Machine with Polynomial Features
svm_model = SVC()
svm_model.fit(train_X_poly, train_y)
pred_svm = svm_model.predict(test_X_poly)
print("SVM with Poly Features Accuracy:", accuracy_score(test_y, pred_svm))
```

```python
# 2. Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(train_X, train_y)
pred_rf = rf_model.predict(test_X)
print("Random Forest Accuracy:", accuracy_score(test_y, pred_rf))
```

```python
# 3. XGBoost Classifier
from sklearn.preprocessing import LabelEncoder
# Encode the target variable
le = LabelEncoder()
train_y_encoded = le.fit_transform(train_y)
test_y_encoded = le.transform(test_y)
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
# Fit the model with the encoded target variable
xgb_model.fit(train_X, train_y_encoded)
pred_xgb = xgb_model.predict(test_X)
# The predictions from XGBoost will be numerical, convert them back to original labels for accuracy calculation
pred_xgb_decoded = le.inverse_transform(pred_xgb)
print("XGBoost Accuracy:", accuracy_score(test_y, pred_xgb_decoded))
```

```python
# 4. Hyperparameter Tuning (SVM Example)
param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
grid = GridSearchCV(SVC(), param_grid, cv=5)
grid.fit(train_X_scaled, train_y)
best_svm = grid.best_estimator_
pred_grid_svm = best_svm.predict(test_X_scaled)
print("Tuned SVM Accuracy:", accuracy_score(test_y, pred_grid_svm))
```

```python
# 5. Cross Validation (SVM)
cross_scores = cross_val_score(SVC(kernel='rbf', C=1), X, y, cv=5)
print("Cross-validated Accuracy (SVM):", cross_scores.mean())
```

```python
# Classification Report for Best Model
print("\nClassification Report (Best Tuned SVM):")
print(classification_report(test_y, pred_grid_svm))
```

```python
# Confusion Matrix
conf_mat = confusion_matrix(test_y, pred_grid_svm)
sns.heatmap(conf_mat, annot=True, cmap='Blues', fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - Tuned SVM')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.svm import SVC

# Load dataset (with headers already present)
df = pd.read_csv('iris_dataset.csv')
print(df.head())
print(df.describe())

# Check the actual column names in the loaded DataFrame
print("\nDataFrame Columns:")
print(df.columns)

# Pairplot Visualization
# Adjust hue if 'target' column name is different
sns.pairplot(df, hue='target', palette=['red', 'blue', 'green'])
plt.show()

# Features and Target
# Use the correct column names from the DataFrame
# Assuming the column names are 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)' based on common variations
# Replace with the exact names printed from df.columns
X = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
Y = df['target'] # Keep 'target' if it's the correct column name for the species

# Split into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train SVM model
model_svc = SVC()
model_svc.fit(X_train, Y_train)

# Predict
predictions = model_svc.predict(X_test)

# Accuracy
accuracy = accuracy_score(Y_test, predictions)
print("\nMODEL : SUPPORT VECTOR MACHINE")
print(f"Accuracy: {accuracy * 100:.2f}%")

# Show Actual vs Predicted
print("\nACTUAL TARGET\tPREDICTED TARGET")
for actual, predicted in zip(Y_test.values, predictions):
    print(f"{actual}\t\t{predicted}")

# Evaluation Reports
print("\nClassification Report:")
print(classification_report(Y_test, predictions))

print("\nConfusion Matrix:")
print(confusion_matrix(Y_test, predictions))

# Real-time User Input
slen = float(input("\nEnter sepal length: "))
swid = float(input("Enter sepal width: "))
plen = float(input("Enter petal length: "))
pwid = float(input("Enter petal width: "))

# Create new DataFrame for prediction
new_data = pd.DataFrame([[slen, swid, plen, pwid]],
                        # Use the correct column names for the new data DataFrame as well
                        columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])

# Predict new input
predicted_species = model_svc.predict(new_data)[0]
print(f"\nThe species of the flower is: {predicted_species}")
```
